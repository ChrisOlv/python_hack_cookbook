```python
import matplotlib.pyplot as plt
```

# Basics

Making a chart with matplotlib always starts by calling theplt.subplots() function.

The subplots() function returns 2 things: a Figure object (fig) and an Axes object (ax).

Think of the fig as the overall canvas where everything is drawn, and ax as the area or section within that canvas where the actual graph or plot appears.

```python
fig, ax = plt.subplots()
```

AX = Fill the axes

```python
ax.plot([1,2,3,4,5], [2,3,1,4,5])
```

plt.show() is always the last step, allowing to actually draw something on the screen:

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4,5], [2,3,1,4,5]) # plot pour line plot
# ax.scatter([1,2,3,4,5], [2,3,1,4,5]) # scatter pour scatter plot
plt.show()
```

```python
fig, ax = plt.subplots()

x = [
   10, 12, 11, 13,
   15, 17, 12, 16,
   15, 21, 9, 14,
   14, 15, 13
]


ax.hist(x)
plt.show()
```

# Pyplot VS Axes
Matplotlib Basics > Pyplot vs Axes

```python
# interface rapide mais mauvaise pratique :

plt.plot([1,2,3,4,5], [2,3,1,4,5])
plt.show() 
```

```python
# interface avec AXES, la best practice : 
fig, ax = plt.subplots()
ax.plot([1,2,3,4,5], [2,3,1,4,5])

plt.show()
```

# Fig, Axes and Subplots

### add text on chart : 


```python

fig, ax = plt.subplots()

ax.plot([1,2,3,4,5], [2,3,1,4,5])
ax.text(3, 3, "Hello world! ax text")
fig.text(0.2, 0.2, "Hello world! fig text")


plt.show()
```
